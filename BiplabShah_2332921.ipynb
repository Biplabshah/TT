```python
#Biplab Shah/2332921
#A matrix A's determinant, inverse, and multiplication can be calculated using a Python program.
import numpy as np

# asking the user for their input on the matrix values
def get_user_matrix():
    while True:
        try:
            values = []
            for i in range(3):
                row = []
                for j in range(3):
                    value = int(input(f"Enter value ({i+1}, {j+1}): "))
                    row.append(value)
                values.append(row)
            A = np.array(values) # create a multidimensional array formatted into 3x3 matrix
            break
        except ValueError:
            print("Error: enter integers only")
    return A

# the matrix values printed
def print_matrix_values(A):
    print("A =\n", A)
    print("Values in A are: ", A.flatten())

# Determine the matrix's determinant.
def find_determinant(A):
    detA = np.linalg.det(A)
    np.set_printoptions(precision=2, suppress=True)
    print("The determinant of A is: ", detA)

# function to determine the matrix's inverse
def find_inverse(A):
    try:
        invA = np.linalg.inv(A)
        np.set_printoptions(precision=2, suppress=True)
        print("The inverse of A is: \n", invA)
    except np.linalg.LinAlgError:
        print("The matrix A is singular.")

# Count the number of values in the matrix that are more than or equal to 10 with this function
def count_large_values(A, threshold):
    count = np.count_nonzero(A >= threshold)
    print("Number of values which are >= ", threshold, " is: ", count)

# according to question 1, functions to carry out the matrix operations A*A and A*A*A 
def matrix_operations(A):
    A2 = np.dot(A, A)
    A3 = np.dot(A2, A)
    print("A*A =\n", A2)
    print("A*A*A =\n", A3)

def main():
    A = get_user_matrix()
    print_matrix_values(A)
    find_determinant(A)
    find_inverse(A)
    count_large_values(A, 10)
    matrix_operations(A)

def run_program():
    while True:
        main()
        rerun = input("Do you want to use the program again? (Y/N) ").upper()
        if rerun == "N":
            break
    exit_program()

def exit_program():
    print("Program has concluded. Thank you for using it.")

run_program()
```
